```python
# Lets see some specific data columns
DataFrame(nfl_frame,columns=['Team','First NFL Season','GP'])

# What happens if we ask for a column that doesn't exist?
DataFrame(nfl_frame,columns=['Team','First NFL Season','GP','Stadium'])
```

```python
# select columns
print(nfl_frame.Team)
nfl_frame['Team']
```

```python
# We can retrieve rows through indexing
nfl_frame.index = [i for i in 'abcde']

# by index location
print(nfl_frame.iloc[3])

# by index
nfl_frame.loc['e']
```

```python
l = [i for i in nfl_frame.columns]
l.append('Stadium')
nfl_frame = DataFrame(nfl_frame,columns=l)
nfl_frame
```

```python
nfl_frame['Stadium'] = "Levi's Stadium"
print(nfl_frame)
nfl_frame['Stadium'] = np.arange(5)
nfl_frame
```
